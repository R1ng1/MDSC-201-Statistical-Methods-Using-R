# Implement a Simple Linear Regression model to predict mpg based on horse power

```r
# Fitting the linear regression model
lm_model <- lm(mpg ~ horsepower, data = Auto)
```

```r
# Make predictions (example)
new_data <- data.frame(horsepower = c(120, 150))
predictions <- predict(lm_model, newdata = new_data)
```

```r
# Printing the predictions
predictions
```

# Display the summary of the regression model and interpreting the significance of the predictor

```r
summary(lm_model)
```

# Interpratation



*   The p-value for horsepower is displayed in the summary output

*   If it is less than our significance level (typically 0.05), then the predictor is statistically significant.
*   This suggests there is a significant linear relationship between horsepower and mpg.

```r
ggplot(Auto, aes(x = horsepower, y = mpg)) +
  geom_point() +                                                                # Adding points for each observation
  geom_smooth(method = "lm", se = FALSE, color = "red") +                       # Adding the regression line
  labs(title = "Scatter Plot of mpg vs. horsepower with Regression Line",
       x = "Horsepower",
       y = "Miles per Gallon")
```

# Convert mpg into binary categorical variable:
*   If mpg > median(mpg) assign it to high efficiency(1).

*   Otherwise assign it to low efficiency(0).



```r
# Calculate the median mpg
median_mpg <- median(Auto$mpg)

# Create the binary categorical variable
Auto$mpg_efficiency <- ifelse(Auto$mpg > median_mpg, 1, 0)

# Display the updated dataset
tail(Auto)
```

# Implementing a Logistic Regression Model to classify fuel efficiency using weight

```r

# Fit the logistic regression model
logistic_model <- glm(mpg_efficiency ~ weight, data = Auto, family = "binomial")
```

```r
# Make predictions (example)
new_data <- data.frame(weight = c(2000, 3000))
predictions <- predict(logistic_model, newdata = new_data, type = "response")
```

```r
# Print the predicted probabilities
predictions
```

# Display the model summary and interpret the coefficients

```r
# Summarize the model
summary(logistic_model)
```


# Interpretation of the Logistic Regression Model



*  1. Model Fit: The model shows a statistically significant relationship between weight and fuel efficiency (mpg_efficiency).  The low residual deviance compared to null deviance and a significant p-value for weight suggest a good model fit.

*  2. Intercept:
The intercept (11.76) represents the log-odds of high fuel efficiency when the weight is zero.  This value, however, is not practically interpretable in this context as a zero weight is not realistic.

*   3. Weight Coefficient:The coefficient for weight (-0.0041) indicates that for every one unit increase in weight, the log-odds of high fuel efficiency decreases by 0.0041.  Since this is log-odds, it's more insightful to exponentiate this value.

*    4. Exponentiated Weight Coefficient:  exp(-0.004119)  results in approximately 0.996. This indicates that for each one unit increase in weight, the odds of high fuel efficiency are multiplied by 0.996 (almost 1). This implies a small negative relationship between weight and the odds of high fuel efficiency.  In other words, heavier cars have slightly lower odds of being classified as high efficiency.


*   5. Statistical Significance: The p-value for weight (< 2e-16) is highly significant. This strongly supports the conclusion that weight is a statistically significant predictor of fuel efficiency,  reinforcing the relationship established in the model.

# Sigmoid Function

```r
# Define a sigmoid function
sigmoid <- function(x) {
  1 / (1 + exp(-x))
}
```

```r
# Example values
x_values <- seq(-5, 5, by = 0.1)
y_values <- sigmoid(x_values)
```

```r
# Plot the sigmoid function (requires ggplot2)
library(ggplot2)
ggplot() +
  geom_line(aes(x = x_values, y = y_values)) +
  labs(title = "Sigmoid Function", x = "x", y = "sigmoid(x)")
```

# Interpretation:



*  The coefficients in the summary output represent the log-odds of high fuel efficiency for a one-unit increase in weight.
  

*  The p-value for weight indicates its statistical significance in predicting fuel efficiency.  
* A low p-value suggests a statistically significant relationship between weight and fuel efficiency.

The predicted probabilities show the estimated probability of high fuel efficiency for the specified weights in new_data.
